## Clase cientifico

```python
# --- INVESTIGACIÓN POO: CLASE CIENTÍFICO ---
class Cientifico:
    def __init__(self, nombre, clave, especialidad):
        self.nombre = nombre
        self.clave = clave
        self.especialidad = especialidad

    def investiga(self):
        print(f"El científico {self.nombre} está investigando nuevas teorías en el área de {self.especialidad} 🔍🔍🔍.")

    def experimenta(self):
        # Aquí se imprime la clave que ahora es un número entero
        print(f"Iniciando experimento... Código de acceso: {self.clave}. Especialidad: {self.especialidad}.🥼🥼🥼")

    def informa(self):
        print(f"El Dr./Dra. {self.nombre} ha publicado los resultados de su investigación.\n👨‍🔬👨‍🔬👨‍🔬")

# --- INSTANCIACIÓN INTERACTIVA CON INPUTS ---

print("--- REGISTRO DE DATOS: CIENTÍFICO 1 ---")
nombre1 = input("Ingrese el nombre: ")
clave1 = int(input("Ingrese la clave (solo números): "))
especialidad1 = input("Ingrese la especialidad: ")
cientifico1 = Cientifico(nombre1, clave1, especialidad1)

print("\n--- REGISTRO DE DATOS: CIENTÍFICO 2 ---")
nombre2 = input("Ingrese el nombre: ")
clave2 = int(input("Ingrese la clave (solo números): "))
especialidad2 = input("Ingrese la especialidad: ")
cientifico2 = Cientifico(nombre2, clave2, especialidad2)

# --- LLAMADAS A LAS FUNCIONES PARA CADA OBJETO ---

print("\n--- Actividades del Científico 1 ---")
cientifico1.investiga()
cientifico1.experimenta()
cientifico1.informa()

print("--- Actividades del Científico 2 ---")
cientifico2.investiga()
cientifico2.experimenta()
cientifico2.informa()

print("Programa terminado")
print("Investigación POO - Adaptado del modelo de Samuel Martinez")
```
